```python
import pandas as pd

# Import CSV files
df_random = pd.read_csv('1_evaluation_report_random.csv')
df_dtw_dbscan = pd.read_csv('2_evaluation_report_dtw_dbscan.csv')
df_dtw_agglomerative = pd.read_csv('3_evaluation_report_dtw_agglomerative.csv')
df_features_agglomerative = pd.read_csv('4_evaluation_report_features_agglomerative.csv')
df_evaluation = pd.read_csv('evaluation_report_filtering_dtw.csv')
```

```python
import matplotlib.pyplot as plt

def plot_boxplot(dfs, names, features, titles):
    """
    Plots box plots for the specified features in multiple dataframes, each feature in a separate figure.

    Parameters:
    dfs (list of pd.DataFrame): A list of dataframes containing the data.
    features (list): A list of feature names to plot.
    """
    num_dfs = len(dfs)

    for i, feature in enumerate(features):
        fig, ax = plt.subplots(figsize=(10, 5))
        data = [df[feature] for df in dfs]
        ax.boxplot(data, tick_labels=[names[i] for i in range(num_dfs)])
        ax.set_title(titles[i])
        plt.tight_layout()
        plt.show()

# Example usage:
# plot_boxplot([df_random, df_features_agglomerative, df_dtw_agglomerative, df_dtw_dbscan], ['Random', 'Features Agglomerative', 'DTW Agglomerative', 'DTW DBSCAN'], ['diversity', 'time_to_fault_ratio', 'fault_to_selection_ratio'])
```

```python
plot_boxplot(
    [df_random, df_features_agglomerative, df_dtw_agglomerative, df_dtw_dbscan, df_evaluation],
    ['Random', 'Features Agglomerative', 'DTW Agglomerative', 'DTW DBSCAN', "Filtering + DTW Agglomerative"],
    ['diversity', 'time_to_fault_ratio', 'fault_to_selection_ratio'],
    ["Diversity", "Time to Fault Ratio", "Fault to Selection Ratio"]
)
```
